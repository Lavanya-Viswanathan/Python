```python
ax=sns.countplot(x="Gender",data=df)
for value in ax.containers:
    ax.bar_label(value,fontsize=10)
    
plt.title("Number of Male and Female in dataset",color="Green",fontsize=15 )
plt.xlabel("Gender",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
ax=sns.countplot(x="Occupation",data=df)
for value in ax.containers:
    ax.bar_label(value,color="red",fontsize=10)
plt.title("Number based on occupation",color="Green",fontsize=15 )
plt.xlabel("occupation",color="blue",fontsize=15)
plt.xticks(rotation=60,color="red")
plt.yticks(color="red")
plt.ylabel("Count",color="blue",fontsize=15)
plt.show()
    
```

```python
sales_wrt_age=df.groupby(['Age Group'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_wrt_age
```

```python
sns.barplot(x="Age Group",y="Amount",data=sales_wrt_age)
plt.title("Sales with respect to Age group",color="red",fontsize=15 )

plt.show()
```

```python
df.columns
```

```python
sns.set(rc={"figure.figsize":(15,7)})
ax=sns.barplot(x="Age Group",y="Amount",data=sales_wrt_age)
for value in ax.containers:
    ax.bar_label(value,color="red",fontsize=10,rotation=60)
plt.title("Salaes With Respect to Age group",color="Green",fontsize=15 )
plt.xlabel("Age_group",color="blue",fontsize=15)
plt.xticks(rotation=60,color="red")
plt.yticks(color="red")
plt.ylabel("Total Amount",color="blue",fontsize=15)
plt.show()
```

```python
sns.set(rc={"figure.figsize":(6,7)})
ax=sns.barplot(x="Gender",y="Amount",data=sales_wrt_gender)
for value in ax.containers:
    ax.bar_label(value,color="red",fontsize=10)
plt.title("Salaes With Respect to Gender",color="Green",fontsize=15 )
plt.xlabel("Age_group",color="blue",fontsize=15)
plt.xticks(color="red")
plt.yticks(color="red")
plt.ylabel("Total Amount",color="blue",fontsize=15)
plt.show()
```

```python
orders_wrt_state=df.groupby(['State'],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
orders_wrt_state
```

```python
sns.set(rc={"figure.figsize":(30,15)})
ax=sns.barplot(data=orders_wrt_state,x="State",y="Orders")
plt.title("Order with respect to State",color="red",fontsize=25 )
for bars in ax.containers:
    ax.bar_label(bars,fontsize=20)
plt.xlabel("States",fontsize=25)
plt.ylabel("Number of Orders",fontsize=25)
plt.show()
```

```python
sales_wrt_MaritalStatus=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_wrt_MaritalStatus
```

```python
sns.set(rc={"figure.figsize":(7,7)})
ax=sns.barplot(data=sales_wrt_MaritalStatus,x="Marital_Status",y="Amount",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Amount with respect to Gender",color="red",fontsize=15 )
plt.xlabel("Marital_Status",fontsize=15)
plt.ylabel("Amount",fontsize=15)
plt.show()
```
