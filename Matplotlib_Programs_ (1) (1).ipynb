```python
#SVM Program
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as mtp
from sklearn import svm
#linear data
x=np.array([1,5,1.5,8,1,9,7,8.7,2.3,5.5,7.7,6.1])
y=np.array([2,8,1.8,8,0.6,11,10,9.4,4,3,8.8,7.5])
#show unclassified data
plt.scatter(x,y)
plt.show()
#shapping data for training the model
training_x=np.stack((x,y)).T
training_y=[0,1,0,1,0,1,1,1,0,0,1,1]
#define the model
clf=svm.SVC(kernel='linear')
#train the model
clf.fit(training_x,training_y)
#get the weight values for the linear equation from the trained svm
w=clf.coef_[0]
#get the y_offset for the linear equation
a=-w[0]/w[1]
#make the x-axis space forthe data points
XX=np.linspace(0,13)
#get the y_values to plot the decision  boundary
YY=a*XX-clf.intercept_[0]/w[1]
#plot the decision boundary
mtp.plot(XX,YY,'k-')
#show the plot visually
mtp.scatter(training_x[:,0],training_x[:,1],c=training_y)
mtp.legend()
mtp.show()
```
